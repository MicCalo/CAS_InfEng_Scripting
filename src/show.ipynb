Find the center of Brooklyn

```python
import importlib
import address_tools

importlib.reload(address_tools)
center_of_brookyn = address_tools.brooklyn.center
```

Prepare folium

```python

import math
import pandas as pd
import folium
from folium import plugins

m = folium.Map(
    location=(center_of_brookyn.y, center_of_brookyn.x), 
    zoom_start=12,
    tiles=None # make sure Stamen Toner is default, because its added first below
)
```

Add prices per sqft 

```python
sales_grouped = pd.read_csv('./../data/prepared/sales_grouped.csv')

folium.Choropleth(
    name="average price per gross square feet",
    geo_data='../data/prepared/ZipCodeBoundaries_BrooklynOnly.geojson',
    data=sales_grouped,
    columns=['ZIP CODE', 'PRICE PER GSQ'],
    key_on='feature.properties.ZIPCODE',
    fill_color='RdYlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Price per gross sq ($)"
).add_to(m)
```

Add schools as points

```python
schools = pd.read_csv('./../data/prepared/schools.csv', ',')

radius = 5    
color = 'green'

for _, row in schools.iterrows():
    row['lon']
    if math.isnan(row['lon']):
        continue
    
    

    folium.CircleMarker(
        location=(row['lat'], row['lon']),
        radius=radius,
        color=color,
        fill_color=color,
        fill_opacity=0.6,
        popup='School',
    ).add_to(m)
```

```python
#adding the different 'backgrounds', black/white as default
folium.TileLayer("Stamen Toner", name="Stamen Toner").add_to(m)
folium.TileLayer("Stamen Terrain", name="Stamen Terrain").add_to(m)
folium.TileLayer("Open Street Map", name="Open Street Map").add_to(m)

#adding the possibility to show result in fullscreen mode
folium.plugins.Fullscreen().add_to(m)

# give the user the possibility to swich backgrounds and select the layer
folium.LayerControl().add_to(m)
m
```
